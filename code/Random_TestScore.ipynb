## 1. Random classifier (Ib)

```python
p_Ib = 12. / 52.
n_Ib = 50
fig = plt.figure(figsize=(10,6))
plt.plot(np.arange(n_Ib), sp.stats.binom.pmf(np.arange(n_Ib), n_Ib, p_Ib))
plt.xlabel('$k$', fontsize = 20)
plt.ylabel('$P(k)$', fontsize = 20)
#plt.axvline(sp.stats.binom.mean(50, p_Ib), color = 'Black')
print("Mean:", sp.stats.binom.mean(n_Ib, p_Ib))
print("Percentage score:", sp.stats.binom.mean(n_Ib, p_Ib) / n_Ib)
```
